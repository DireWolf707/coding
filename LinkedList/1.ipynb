```python
class Node:
    def __init__(self, d=None, n=None):
        self.data = d
        self.next = n
        
class List:
    head = None
    tail = None

    def push_front(self,d):
        if self.head is None:
            node = Node(d)
            self.head = self.tail = node
        else:
            node = Node(d, self.head)
            self.head = node

    def push_back(self,d):
        if self.head is None:
            node = Node(d)
            self.head = self.tail = node
        else:
            node = Node(d)
            self.tail.next = node
            self.tail = node

    def print(self):
        curr = self.head
        while(curr):
            print(curr.data,end=' -> ')
            curr = curr.next
        print('None')

    def insert(self,d,pos):
        if pos == 0:
            self.push_front(d)
        else:
            curr = self.head
            while (curr.next and pos > 1):
                curr = curr.next
                pos-=1

            if curr == self.tail:
                node = Node(d,None)
                self.tail = node
            else:
                node = Node(d,curr.next)

            curr.next = node

    def search_i(self,d):
        curr = self.head
        idx = 0
        while (curr):
            if curr.data == d: return idx
            curr = curr.next
            idx += 1
        return -1

    def search_r(self,d):
        def helper_fxn(n,d,idx=0):
            if n is None:
                return -1
            if n.data == d: 
                return idx
            return helper_fxn(n.next,d,idx+1)

        return helper_fxn(self.head,d)

    def destroy(self):
        def helper_fxn(n):
            if n is None:
                return
            helper_fxn(n.next)
            n.next = None

        helper_fxn(self.head)
        self.head = None

    def pop_front(self):
        if self.head is None:
            return None
        elif self.head == self.tail:
            temp = self.head.data
            self.head = self.tail = None
            return temp
        else:
            temp = self.head.data
            self.head = self.head.next
            return temp

    def pop_back(self):
        if self.tail is None: # tail and head can only be None together
            return None
        elif self.tail == self.head:
            temp = self.tail.data
            self.tail = self.head = None
            return temp
        else:
            curr = self.head
            while curr.next != self.tail:
                curr = curr.next
            temp = self.tail.data
            curr.next = None
            self.tail = curr
            return temp
    
    def remove(self,pos):
        if pos == 0:
            return self.pop_front()
        curr = self.head
        for i in range(0,pos-1):
            curr = curr.next
        if curr.next == self.tail:
            curr.next = None
            self.tail = curr
        else:
            curr.next = curr.next.next

    def reverse(self):
        prevNode = None
        currNode = self.head
        while currNode != None:
            nextNode = currNode.next

            currNode.next = prevNode
            prevNode = currNode
            currNode = nextNode
        self.head, self.tail = self.tail, self.head
```

```python
l = List()
l.push_back(3)
l.push_front(4)
l.push_back(2)
l.push_back(1)
l.print()
l.reverse()
l.print()
l.reverse()
l.print()
l.reverse()
l.print()
```
